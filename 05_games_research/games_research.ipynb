# Исследование продаж компьютерных игр

Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Данные по 2016 год включительно. В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

# Цели исследования

Выявить определяющие успешность игры закономерности.

Проверить гипотезы:
- Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
- Средние пользовательские рейтинги жанров Action и Sports разные.

# Описание данных
- `Name` — название игры
- `Platform` — платформа
- `Year_of_Release` — год выпуска
- `Genre` — жанр игры

```python
# напишем функцию, которая выводит диаграмму 
# с долями пропусков по убыванию

def pass_value_barh(data):
    try:
        (
            (data.isna().mean()*100)
            .to_frame()
            .rename(columns = {0:'space'})
            .query('space > 0')
            .sort_values(by='space', ascending=True)
            .plot(kind = 'barh', figsize=(7, 5), legend=False, fontsize = 15)
            .set_title('Доля пропущенных значений, %' + "\n", fontsize = 15)
        );
    except:
        print('пропусков не осталось')
```

### Обработка пропусков

**Год выпуска** 

Можно попробовать заполнить пропуски, используя группировку по названию. Но сначала проверим отличаются ли годы выпуска одних и тех же игр на разных платформах.

```python
# создадим таблицу с годом выпуска с группировкой по названию и платформе

name_platform_year = data.groupby(['name', 'platform']).agg({'year_of_release':'first'}).reset_index()
```

```python
# создадим таблицу с числом платформ для каждой игры

name_platform_count = name_platform_year.groupby('name').agg({'platform':'count'}).\
reset_index().rename(columns={'platform':'count'})
```

```python
# объединим таблицы

name_platform_year = name_platform_year.merge(name_platform_count, on='name', how='left')
```

```python
# отберем строки, где от 2-х платформ,
# сгруппируем по названию и платформе 
# и посмотим на годы выпуска (первые 50 строк)

name_platform_year.query('count >= 2').groupby(['name', 'platform']).agg({'year_of_release':'first'}).head(50)
```

В срезе попадаются варианты, где для разных платформ годы выпуска отличаются. В основном, если разница есть, то она равна 1 году. Но где-то разница 3 года, а где-то целых 9 лет.

## Портрет пользователей по регионам

Посмотрим на доли 5 самых популярных **платформ** в каждом регионе c 2013 г.. Остальные платформы объединим в группу **Other**.

```python
for r in ['na_sales', 'eu_sales', 'jp_sales']:
    top5 = data_from_2013.groupby('platform').agg({r:'sum'}).\
    sort_values(r, ascending=False).reset_index()['platform'].head()
    
    data_r = data_from_2013
    data_r['platform_group'] = data_r['platform'].\
    where(data_r['platform'].isin(top5), 'Other')
    
    data_r.groupby('platform_group').agg({r:'sum'}).\
    sort_values(r, ascending=False).\
    plot(y=r, kind='pie', autopct='%.1f', figsize=(6,6));
    plt.legend(bbox_to_anchor=(1, 1));
    
    if r == 'na_sales':
        region = 'Северной Америке'
    elif r == 'eu_sales':
        region = 'Европе'
    elif r == 'jp_sales':
        region = 'Японии'
    plt.title(f'Доли продаж по платформам в {region}');
```

Данные будем брать, начиная с 2013 года, т.к. этот период нам больше подойдет для планирования инвестиций.

**Первая гипотеза.** Сформулируем нулевую и альтернативную гипотезы:

- H₀: средние пользовательские рейтинги платформ **XOne и PC** равны;
- H₁: средние пользовательские рейтинги платформ **XOne и PC** не равны.

Создадим соответствующие массивы и сравним средние оценки пользоватей. Выборки независимы и при этом могут отличаться по размерам. Это повод для того, чтобы в качестве критерия выбрать t-тест для независимых генеральных совокупностей.

```python
data_pc = data_from_2013.\
loc[data['platform']=='PC', 'user_score'].dropna().astype('int')
```

```python
data_xone = data_from_2013.\
loc[data['platform']=='XOne', 'user_score'].dropna().astype('int')
```

```python
alpha = 0.05
results = st.ttest_ind(data_pc, data_xone, equal_var=False)
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

По итогам теста, мы можем сказать, что значимых различий между средними пользовательскими оценками платформ **XOne и PC** нет.

**Вторая гипотиза.** Сформулируем нулевую и альтернативную гипотезы:

- H₀: средние пользовательские рейтинги жанров **Action и Sports** *не равны*;
- H₁: средние пользовательские рейтинги жанров **Action и Sports** равны.

Опять же выборки независимы и отличаются по размерам: применяем t-тест для независимых генеральных совокупностей.

```python
data_action = data_from_2013.\
loc[data['genre']=='Action', 'user_score'].dropna().astype('int')
```

```python
data_sports = data_from_2013.\
loc[data['genre']=='Sports', 'user_score'].dropna().astype('int')
```

```python
alpha = 0.05
results = st.ttest_ind(data_action, data_sports, equal_var=False)
print('p-значение:', results.pvalue)

if results.pvalue > alpha:
# знак "больше", т.к нулевая гипотеза утверждает разность средних
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

По итогам теста, мы можем сказать, что средние пользовательские рейтинги жанров **Action и Sports** различны.

**Вывод**

Средние пользовательские оценки платформ **XOne и PC** не имеют значимых различий. А средние пользовательские рейтинги жанров **Action и Sports** отличаются.

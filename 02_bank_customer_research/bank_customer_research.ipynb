# Исследование надежности заемщиков

Заказчик — кредитный отдел банка. Входные данные от банка — статистика о платёжеспособности клиентов.
Результаты исследования будут учтены при построении модели кредитного скоринга — специальной системы, которая оценивает способность потенциального заёмщика вернуть кредит банку.

Нужно разобраться, влияет ли семейное положение и количество детей клиента на факт погашения кредита в срок.

## Предобработка данных

### Удаление пропусков

```python
# посчитаем количество пропусков в данных

data.isna().sum()
```

```python
# посмотрим на первые 20 строк таблицы 
# с пропусками в столбце days_employed

data.query('days_employed.isna()').head(20)
```

В столбцах `days_employed` и `total_income` есть пропущенные значения. И их число одинаково (по 2147). Срез таблицы выше показывает, что пропуски в обеих колонках находятся в одних и тех же строках. Вероятно, произошел сбой при выгрузке данных.    

Пропуски в столбце `days_employed` мы обработаем на следующем этапе. А пока займемся столбцом `total_income`, который хранит данные о доходах. На сумму дохода сильнее всего влияет тип занятости, поэтому заполнить пропуски в этом столбце нужно медианным значением по каждому типу из столбца `income_type`.

```python
for type in data['income_type'].unique():
    data.loc[(data['income_type'] == type) & (data['total_income'].isna()), 'total_income'] = \
    data.loc[(data['income_type'] == type), 'total_income'].median()
```

### Обработка аномальных значений

В столбце `days_employed` попадаются отрицательные значения количества дней трудового стажа. Заменим все отрицательные значения положительными с помощью метода `abs()`.

```python
data['days_employed'] = data['days_employed'].abs()
```

Для каждого типа занятости выведем медианное значение трудового стажа `days_employed` в днях.

```python
data.groupby('income_type')['days_employed'].median()
```

У двух типов (безработные и пенсионеры) получатся аномально большие значения. Исправить такие значения сложно, поэтому оставим их как есть. Тем более этот столбец не понадобится нам для исследования.

```python
# выведем перечень уникальных значений столбца children

data['children'].unique()
```

В столбце `children` есть два аномальных значения. Удалим строки, в которых встречаются такие аномальные значения из датафрейма `data`.

```python
data = data.loc[(data['children'] >= 0) & (data['children'] != 20)]
```

Ещё раз выведем перечень уникальных значений столбца `children`, чтобы убедиться, что аномалии удалены.

```python
data['children'].unique()
```

### Удаление пропусков (продолжение)

Заполним пропуски в столбце `days_employed` медианными значениями по каждому типу занятости `income_type`.

```python
for type in data['income_type'].unique():
    data.loc[(data['income_type'] == type) & (data['days_employed'].isna()), 'days_employed'] = \
    data.loc[(data['income_type'] == type), 'days_employed'].median()
```

```python
# убедимся, что все пропуски заполнены

data.isna().sum()
```

### Изменение типов данных

```python
# заменим вещественный тип данных в столбце total_income на целочисленный

data['total_income'] = data['total_income'].astype('int')
```

### Обработка дубликатов

Обработаем неявные дубликаты в столбце `education`. В этом столбце есть одни и те же значения, но записанные по-разному: с использованием заглавных и строчных букв. Приведем их к нижнему регистру.

```python
data['education'] = data['education'].str.lower()
```

```python
# выведем на экран количество строк-дубликатов в данных

data.duplicated().sum()
```

```python
# удалим дубликаты

data = data.drop_duplicates()
```

### Категоризация данных

На основании диапазонов, указанных ниже, создадим в датафрейме `data` столбец `total_income_category` с категориями:

- 0–30000 — `'E'`;
- 30001–50000 — `'D'`;
- 50001–200000 — `'C'`;
- 200001–1000000 — `'B'`;
- 1000001 и выше — `'A'`.

```python
def categorize_income(income):
    if income <= 30000:
        return 'E'
    elif 30001 <= income <= 50000:
        return 'D'
    elif 50001 <= income <= 200000:
        return 'C'
    elif 200001 <= income <= 1000000:
        return 'B'
    else:
        return 'A'
```

```python
data['total_income_category'] = data['total_income'].apply(categorize_income)
```

Выведем на экран перечень уникальных целей взятия кредита из столбца `purpose`.

```python
data['purpose'].unique()
```

Создадим функцию, которая на основании данных из столбца `purpose` сформирует новый столбец `purpose_category`, в который войдут следующие категории:**

- `'операции с автомобилем'`,
- `'операции с недвижимостью'`,
- `'проведение свадьбы'`,
- `'получение образования'`.

```python
def categorize_purpose(purpose):
    if 'авто' in purpose:
        return 'операции с автомобилем'
    elif 'недвиж' in purpose or 'жил' in purpose:
        return 'операции с недвижимостью'
    elif 'свад' in purpose:
        return 'проведение свадьбы'
    elif 'образов' in purpose:
        return 'получение образования'
    else:
        return 'другое'
```

```python
# добавим колонку с категорией цели кредита в таблицу

data['purpose_category'] = data['purpose'].apply(categorize_purpose)
```

## Зависимость возврата кредита в срок от других параметров

###  Зависимость от количества детей

```python
# посчитаем число заемщиков и доли задолженностей
# по категориям числа детей
# для подсчета доли можем использовать функцию mean(),
# т.к. значения колонки debt либо 1, либо 0

data.pivot_table(index='children', values='debt', aggfunc=['count', 'mean']).reset_index()
```

Категории заемщиков с 4-5 детьми слишком маленькие. Это может исказить выводы из-за недостатка данных. Логично будет выделить более крупную категорию "многодетные", включающую 3 и более детей.

Напишем функцию, которая присвоит категорию в зависимости от числа детей.

```python
def children_category(children):
    try:
        if children == 0:
            return 'бездетные'
        elif children == 1:
            return '1 ребенок'
        elif children == 2:
            return '2 ребенка'
        elif children >= 3:
            return 'многодетные'
    except:
        return 'нет категории'
```

```python
# добавим в таблицу столбец children_category,
# применив созданную функцию

data['children_category'] = data['children'].apply(children_category)
```

```python
# посчитаем число заемщиков и доли задолженностей
# по новым категориям числа детей

data.groupby('children_category').agg({'debt':['count', 'mean']}).reset_index()
```

**Выводы** 

В целом заметно, что наименьшая доля задолженностей у бездетных заемщиков. Она составляет 7,5%. Наибольший процент долгов у категории с 4 детьми (9,7%). Однако категории с количеством детей 4-5 достаточно маленькие, их сложно считать репрезентативными. Нужно констатировать, что выборка не сбалансирована, классы в выборке имеют разные размеры.

Возможно, в рамках имеющихся данных сделать вывод о том, что семьи с детьми относительно семей без детей имеют какие-то задержки в выплатах кредитов, нельзя. Слишком уж близки доли, чуть изменится выборка и значения могут измениться в другую сторону. То есть можно сделать вывод, что сколь угодно значимой разницы между группами нет. Однако, возможно, что даже небольшая разница может существенно отразиться на кредитном портфеле банка.

Но если считать, что любая разница между группами есть повод для «разделения» классов, то можно сделать такой вывод: кредитополучатели без детей являются более ответственными, чем клиенты, у которых есть дети. Видимо, наличие ребенка создает для клиентов дополнительную финансовую нагрузку и клиенты либо менее ответственно относятся к своим кредитным обязательствам, либо не имеют возможности в срок оплачивать обязательства.

После объединения категорий 3, 4, 5 детей в категорию "многодетные" видно, что у них доля задолженностей (8,2%) даже меньше, чем у заемщиков с 1-2 детьми (9,2% и 9,5% соответственно). Нам неизвестно, что было бы, если бы выборка была сбалансированной. Сейчас на долю задолженностей категории "многодетные" сильно повлиял этот показатель в категории 3 ребенка.  Не исключено искажение реальной картины из-за недостатка данных. Но если бы тенденция сохранилась и при сбалансированной выборке, можно былро бы предположить, что 3 и более детей имеют более обеспеченные и ответственные заемщики. Или же многодетные семьи имеют большую финансовую поддержку со стороны государства, что облегчает им выплату кредитов или увеличивает первоначаьный взнос. Естественно, возможны и иные причины.

### Зависимость от семейного положения

```python
# посчитаем число заемщиков и доли задолженностей
# по категориям семейного положения

data.groupby('family_status').agg({'debt':['count', 'mean']}).reset_index()
```

**Выводы** 

Наименьшая доля задолженностей у категории "вдовец/вдова" (6,6%). Однако это самая малочисленная категория, что может вызывать сомнения по поводу реальной частоты возникновения задолженностей у них. Наибольшая - у "Не женат/не замужем" (9,8%). На втором месте по доле долгов заемщики в гражданском браке (9,3%). Категории "женат/замужем" и "в разводе" имеют доли задолженностей 7,6% и 7% соответственно.

С уже озвученными ранее оговорками (если закрыть глаза на различия размеров классов и считать любую разницу между ними поводом для их «разделения»), можем отметить, что банку следует осторожнее относиться к клиентам, которые не состоят или не состояли в браке. Согласно данным такие клиенты менее ответственные. Получается, что официальные отношения (включая и бывшие отношения) ведут к более ответственному отношению к своим обязательствам. Это совпадает со здравым смыслом - семейные люди более щепетильно берегут свое семейное спокойствие и несут за него ответственность. 

### Зависимость от уровня дохода

```python
# посчитаем число заемщиков и доли задолженностей
# по категориям дохода

data.groupby('total_income_category').agg({'debt':['count', 'mean']}).reset_index()
```

Напомним категории дохода:

- 0–30000 — `'E'`;
- 30001–50000 — `'D'`;
- 50001–200000 — `'C'`;
- 200001–1000000 — `'B'`;
- 1000001 и выше — `'A'`.

Из полученной таблицы видно, что категории A, D, E слишком малы, а значит, нерепрезентативны. 
Логично разделить все займы на 2 новые категории по показателю дохода:

- 0–200000 — `'B'`;
- 200001 и выше — `'A'`.

Для этого напишем соответствующую функцию.

```python
def dual_categorize_income(income):
    try:
        if 0 <= income <= 200000:
            return 'B'
        elif income >= 200001:
            return 'A'
    except:
        pass
```

```python
# добавим в таблицу колонку с новой категоризацией по доходу

data['dual_income_category'] = data['total_income'].apply(dual_categorize_income)
```

```python
# посчитаем число заемщиков и доли задолженностей
# по новым категориям дохода

data.groupby('dual_income_category').agg({'debt':['count', 'mean']}).reset_index()
```

**Выводы** 

Чаще всего должниками становятся люди с наименьшим доходом (категория E). Однако интересно, что с ростом доходов тенденция (D->C и B->А) к снижению числа задолженностей не сохраняется, а скорее даже растёт. Возможно люди "перестают считать деньги" или планировать наперёд. Это не позволяет назвать уровень дохода надёжным критерием при определении платёжеспособности клиента. Но может это искажение, вызванное недостаточным размером выборки клиентов категории доходов А.

Если посмотреть на более укрупненные группы, то здесь уже всё более логично. Заемщики с бОльшим доходом (новая категория A, 200001 и выше) ожидаемо имеют меньшую долю задолженности по займам: 7% против 8,5% для категории заемщиков с доходом до 200000 (категория B). Однако категория B более чем в 3 раза превосходит по численности категорию A. Это говорит о том, что около 70% заемщиков имеют доходы до 200000. На основе этих данных сложно приоритизировать одну из двух категорий. Большое количество пропусков и сильно отличающиеся размеры групп не дают возможность сделать однозначный вывод.

### Зависимость от целей кредита

```python
# посчитаем число заемщиков и доли задолженностей
# по категориям целей кредита

data.groupby('purpose_category').agg({'debt':['count', 'mean']}).reset_index()
```

**Выводы** 

Наибольшие доли задолженности по займам у категорий "операции с автомобилем" и "получение образования" (по 9,3% для каждой категории). С авто вроде закономерно: люди покупают в кредит гораздо более дорогие авто, чем те, которые соответствуют их текущим доходам. Странно, что по категории "образование" тоже относительно высокий уровень задолженности: такие кредиты обычно берут люди мотивированные, имеющие впереди цель. Выдача займов для операций с недвижимостью (наиболее популярной по целям займа среди заемщиков) является самой безопасной с точки зрения рисков возникновения задолженности - 7,3%. Чуть больше процент задолжненностей по категории "проведение свадьбы" - 7,9%. Предположительно, такой показатель у категории "свадьба" может быть связан с традицией дарить на свадьбы крупные суммы денег, которые заемщики могут направить на погашение задолженности.

По семейному положению:
- заемщики, которые не состоят или ранее не состояли в официальном браке, имеют наибольшие доли задолдженности (от 9,3% до 9,8%).
- наименьшая доля задолженностей у категории "вдовец/вдова" (6,6%);
- заемщики в браке и в разводе имеют доли задолженности 7,6% и 7% соответственно. 

По ежемесячному доходу:
- заемщики с доходом до 200000 имеют задолженность в 8,5% случаев;
- заемщики с доходом более 200000 имеют задолженность в 7% случаев;

По цели займа:
- меньше всего доли задолженности по займам для операций с недвижимостью - 7,3%;
- больше всего - в категориях "операции с автомобилем" и "получение образования" (по 9,3%);
- 7,9% в категории "проведение свадьбы".
